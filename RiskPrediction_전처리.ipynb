PROC SORT DATA=S.gy40 OUT=CVD NODUPKEY;
BY KEY_SEQ;
WHERE (SUBSTR(SICK_SYM, 1, 3) IN ('I20' 'I21' 'I22' 'I23' 'I24' 'I25') OR SUBSTR(SICK_SYM, 1, 2) IN ('I6')) AND SICK_DIV_TYPE_CD^='3';RUN;

PROC SORT DATA=S.gy20 OUT=gy20 NODUPKEY;
BY KEY_SEQ;
WHERE FORM_CD='02' AND VSCN>=2; RUN;

DATA CVD;
MERGE CVD(KEEP=KEY_SEQ IN=IN1) gy20(IN=IN2 KEEP=PERSON_ID RECU_FR_DT KEY_SEQ);
BY KEY_SEQ;
IF IN1 AND IN2;RUN;

PROC SORT DATA=CVD NODUPKEY;BY PERSON_ID RECU_FR_DT;RUN;
PROC SORT DATA=CVD NODUPKEY;BY PERSON_ID ;RUN;

DATA CVD; SET CVD ;
FORMAT CVD_DT YYMMDD10.;
CVD_DT = INPUT(RECU_FR_DT, YYMMDD10. );
RUN;

/** before diagsnosis or prescription of Hypertension, CVD event가 발생 환자는 제외 **/
DATA POP_FINAL;

BY PERSON_ID;
IF IN1;
IF .< CVD_DT <=INPUT(RECU_FR_DT, YYMMDD10.) THEN DELETE; RUN;

/* 건강검진 정보 */

/* gy20 : 명세서 일반 내역 
   gy30 : 진료 내역
   gy40 : 상병 내역
   gy60 : 처방전 세부 내역 */
PROC SORT DATA = S.gy20 OUT = VISIT1;
BY PERSON_ID KEY_SEQ ; RUN;
PROC SORT DATA = S.gy30 OUT = VISIT2;
BY PERSON_ID KEY_SEQ ; RUN;

PROC SORT DATA = S.gy60 OUT = VISIT3;
BY PERSON_ID KEY_SEQ ; RUN;

DATA S.VISIT1(KEEP=PERSON_ID KEY_SEQ RECU_FR_DT FORM_CD MAIN_SICK SUB_SICK) ;

BY PERSON_ID; IF IN2; RUN;

````python
visit1 = pd.read_csv("./visit1.csv")
display(visit1.head())
print(visit1.shape)
````

````python
# 제외 :  ['A', 'B', 'C', 'L', 'P', 'V', 'W', 'X', 'Y', 'Z']
visit1.loc[visit1['MAIN_SICK'].str[0:1].isin(['A', 'B', 'C', 'L', 'P', 'V', 'W', 'X', 'Y', 'Z']), 'MAIN_SICK'] = '_'
visit1.loc[visit1['SUB_SICK'].str[0:1].isin(['A', 'B', 'C', 'L', 'P', 'V', 'W', 'X', 'Y', 'Z']), 'SUB_SICK'] = '_'

# 주, 부상병 모두 활용
visit1['SICK1'] = 'SICK_' + visit1['MAIN_SICK'].str[0:2]
visit1['SICK2'] = 'SICK_' + visit1['SUB_SICK'].str[0:2]
visit1.head()
````

````python
SICK1 = visit1.groupby('PERSON_ID')['SICK1']
SICK1_CNT = SICK1.value_counts().reset_index(name='counts')

SICK2 = visit1.groupby('PERSON_ID')['SICK2']
SICK2_CNT = SICK2.value_counts(dropna=True).reset_index(name='counts')

display(SICK1_CNT.head())
display(SICK2_CNT.head())
````

````python
SICK_CD = list( set(visit1['SICK1'].dropna()) |  set(visit1['SICK2'].dropna() ) )

print(len(SICK_CD))
print(SICK_CD)

for sick in SICK_CD :
  pop[sick]=0

display(pop.head())
display(pop.shape)
````

````python
# 고혈압 발생하기 전의 데이터로..
for i, person in enumerate(pop['PERSON_ID']) : 
  tmp1 = SICK1_CNT[SICK1_CNT['PERSON_ID']==person]
  tmp2 = SICK2_CNT[SICK2_CNT['PERSON_ID']==person]
  pop.loc[i, tmp1['SICK1'].values ] = list(1 * (tmp1['counts'].values > 0 ) ) 
  pop.loc[i, tmp2['SICK2'].values ] = list(1 * (tmp2['counts'].values > 0 ) )
  
````

````python
pop= pop.drop(['SICK__'], axis=1)
pop.head()
````

````python
visit2 = pd.read_csv("./visit2.csv")
visit3 = pd.read_csv("./visit3.csv")

ATC = pd.read_csv("./ATC.csv", engine='python')
ATC2 = ATC.iloc[:,[2,6,7]]
ATC2.columns = ['GNL_NM_CD', 'ATC', 'NAME']

display(visit2.head())
display(ATC2.head())

visit2 = pd.merge(visit2, ATC2, how='left', on='GNL_NM_CD')
visit2.head()
display(visit2.head())

visit3 = pd.merge(visit3, ATC2, how='left', on='GNL_NM_CD')
visit3.head()
display(visit3.head())
````

````python
# 약 처치를 제외한 처치 코드
# GNL_NM_CD 가 NaN 인데 DIV_CD 가 기록된 경우
visit2.loc[visit2['GNL_NM_CD'].isna(), 'DIV_CD1'] = 'PROC_' + visit2.loc[visit2['GNL_NM_CD'].isna(), 'DIV_CD'].str[0:2]
display(visit2.head())

DIV_CD = list( set(visit2['DIV_CD1'].dropna() ) )
print(len(DIV_CD))
print(DIV_CD)

for div in DIV_CD :
  pop[DIV_CD]=0

display(pop.head())
display(pop.shape)
````

````python
DIV = visit2.groupby('PERSON_ID')['DIV_CD1']
DIV_CNT = DIV.value_counts().reset_index(name='counts')
display(DIV_CNT.head())

for i, person in enumerate(pop['PERSON_ID']) : 
  tmp = DIV_CNT[DIV_CNT['PERSON_ID']==person]
  pop.loc[i, tmp['DIV_CD1'].values ] = list(1 * (tmp['counts'].values > 0 ) ) 

# 해석 불가 및 필요 없는 코드 삭제
#pop=pop.drop(['PROC_3', 'PROC_AA', 'PROC_15', 'PROC_10', PROC_AL'], axis=1)
pop.head()
````

````python
# 원내, 원외의 약 처치
visit2['GNL_CD'] = 'DRUG_' + visit2['ATC'].str[0:4]
display(visit2.head())
visit3['GNL_CD'] = 'DRUG_' + visit3['ATC'].str[0:4]
display(visit3.head())

GNL_CD = list( set(visit2['GNL_CD'].dropna()) | set( visit3['GNL_CD'].dropna() ) )
print(len(GNL_CD))
print(GNL_CD)

for gnl in GNL_CD :
  pop[gnl]=0

display(pop.head())
display(pop.shape)
````

````python
GNL1 = visit2.groupby('PERSON_ID')['GNL_CD']
GNL2 = visit3.groupby('PERSON_ID')['GNL_CD']

GNL_CNT1 = GNL1.value_counts().reset_index(name='counts')
GNL_CNT2 = GNL2.value_counts().reset_index(name='counts')

for i, person in enumerate(pop['PERSON_ID']) : 
  tmp1 = GNL_CNT1[GNL_CNT1['PERSON_ID']==person]
  tmp2 = GNL_CNT2[GNL_CNT2['PERSON_ID']==person]

  pop.loc[i, tmp1['GNL_CD'].values ] = list(1 * (tmp1['counts'].values > 0 ) ) 
  pop.loc[i, tmp2['GNL_CD'].values ] = list(1 * (tmp2['counts'].values > 0 ) ) 
  
pop = pop.drop(['DRUG_-'], axis=1)
pop.head()
````
